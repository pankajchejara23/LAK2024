# CoTrack

CoTrack is a web-application which allows researchers to collect multimodal data from group activities using device's inbuilt microphone and camera. The application comes with an easy-to-use interface which makes it easier for non-technical MMLA users (e.g., teachers) to collect multimodal data. 



## Technical Details of CoTrack
CoTrack has been developed using Django framework. It uses [Etherpad](https://etherpad.org/) to offer a collaborative writing space to group's participants. To track the participants' writing activities, CoTrack uses Etherpad's REST API to fetch writing logs. 

Additionally, CoTrack allows users to customize what additional data i.e., audio/video they want to record/preprocess. In the case of recording, the multimedia files (e.g., audio/video) are saved at server in WAV/MKV format. These files are saved for the entire duration of the group activity. To avoid any data loss issue due to network problems, CoTrack also save multimedia files in chunks (for every 5 minutes) and send it to the server. In the case of preprocessing, CoTrack does not record audio or video data but instead performs real-time feature extraction and save those features on the server.



CoTrack performs Voice Activity Detection (VAD) and Speech-to-Text conversion in real-time to extract features such as speaking time, number of turns taken, and speech data. These feature extraction takes place on the client's side. Therefore, if the teacher has configured CoTrack only to extract features only then the audio/video data never leaves the user's device.
